```python
def scrap_city(city, url, page=1):
  print(f"Scraping halaman {page} untuk kota {city}")

  header = {
      "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
  }
  # get the HTML content
  response = requests.get(url + f"?page={page}", headers=header)

  # check response
  # return(response)

  # use parser
  soup = BeautifulSoup(response.text, "html.parser")

  # get the main container
  container = soup.find('div', {'class': 'card-list-section'})
  #get the house element
  featured = container.find_all('div', {'class': 'featured-card-component'})

  print(f"Jumlah rumah: {len(featured)}")

  data = []
  for idx, house in enumerate(featured):
    print(f"Scraping rumah ke-{idx + 1}")
    content = house.find('div', {'class': 'card-featured__middle-section'})
    if content is None:
      continue
    price = content.find('div', {'class': 'card-featured__middle-section__price'}).text.strip()
    print(price)
    title = content.contents[2].text.strip()
    location = content.contents[3].text.strip()
    # print(title)
    # print(location)

    features = content.find('div', {'class': 'card-featured__middle-section__attribute'}).text.strip()
    attribute = features.find_all('span', {'class': 'attribute-text'})

    bedrooms = attribute[0].text.strip()
    bathrooms = attribute[1].text.strip()
    garage = attribute[2].text.strip() if len(attribute) > 2 else ""

    area = features.contents[1].text.strip() # 2nd child
    building_area = features.contents[2].text.strip() # 3rd child

    data.append({
        'city': city,
        'title': title,
        'price': price,
        'location': location,
        'area': area,
        'building_area': building_area,
        'bedrooms': bedrooms,
        'bathrooms': bathrooms,
        'garage': garage
    })

  time.sleep(10) # pause 3s
  return data
```

## **4. Fungsi untuk Mulai Scrap**

Fungsi `start_scrap` berfungsi untuk memulai proses scraping pada beberapa kota. Didalamnya, fungsi ini akan memanggil fungsi `scrap_city`.

```python
def start_scrap(cities, max_page_per_city=10):
  data = []

  for city, url in cities.items():
    for page in range(1, max_page_per_city + 1):
      page_data = scrap_city(city, url, page)
      data.extend(page_data)

  return data
```
